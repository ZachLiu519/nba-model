```python
nba = pd.read_excel('leagueaverages.xlsx')
```

```python
columns = nba.columns.values[3:]
```

```python
interval = nba.iloc[::5, :]
interval
```

```python
plt.plot('Season', '3P', data = nba)
plt.title('3P over Time')
plt.xlabel('Season')
plt.ylabel('3P');
plt.xticks(np.arange(len(nba['Season']), step = 1), nba['Season'], rotation='vertical')

fig = plt.figure()
ax = fig.add_axes([0, 0, 50, 13])
ax.set_xticks([0, 5, 10, 15, 20])
ax.set_xlabels(interval['Season'])
```

```python
nba.plot(x = 'Season', y = '3P', kind = 'line', rot = 90)
```

```python
for name in columns:
    nba.plot(x = 'Season', y = name, kind = 'line', rot = 90)
    plt.ylabel(name)
    plt.title(name + ' over Time')
```
