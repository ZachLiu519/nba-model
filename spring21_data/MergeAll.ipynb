```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# load all datasets

raptor_with_rapm = pd.read_excel('RAPTOR_with_RAPM.xlsx')
lebron = pd.read_excel('LEBRON_15_20.xlsx')
rpm = pd.read_excel('RPM_15_20.xlsx')
```

```python
rpm = rpm.drop(['Unnamed: 0', 'RK', 'RPM', 'WINS'], axis=1)
```

```python
lebron = lebron.drop(['Unnamed: 0', 'bref_id', 'LEBRON'], axis=1)
```

```python
raptor_with_rapm = raptor_with_rapm.drop(['Unnamed: 0', 'season_y', 'teamId', 'primaryKey',
                                         'season_x', 'raptor_total', 'playerId', 'playerName',
                                         'LA_RAPM', 'LA_RAPM_Rank', 'LA_RAPM__Def_Rank', 'LA_RAPM__Off_Rank',
                                         'RA_EFG__Def_Rank', 'RA_EFG__Off_Rank', 'RA_FTR_Rank', 'RA_FTR__Def_Rank',
                                         'RA_FTR__Off_Rank', 'RA_ORBD_Rank', 'RA_EFG_Rank', 'predator_total',
                                         'RA_ORBD__Def_Rank', 'RA_ORBD__Off_Rank', 'RA_TOV_Rank', 'RA_TOV__Def_Rank',
                                         'RA_TOV__Off_Rank', 'RAPM', 'RAPM_Rank', 'RAPM__Def_Rank', 'RAPM__Off_Rank', 'RA_TOV',
                                         'RA_ORBD', 'RA_FTR', 'RA_EFG'], axis=1)
```

```python
raptor_with_rapm.columns
```

```python
raptor_with_rapm_16_20 = raptor_with_rapm.loc[raptor_with_rapm['clean_season'] >= 2016]
```

```python
raptor_with_rapm_16_20
```

```python
rpm_16_20 = rpm.loc[rpm['end season'] < 2021]
```

```python
rpm_16_20
```

```python
pd.options.display.max_columns = 100
```

```python
lebron_rpm = lebron.merge(rpm_16_20, left_on=['Player', 'end season'], right_on=['NAME', 'end season']).drop(['NAME', 'TEAM', 'GP'], axis = 1)
```

```python
all_impact_metric = lebron_rpm.merge(raptor_with_rapm_16_20, left_on=['Player', 'end season'], right_on=['player_name', 'clean_season']).drop(['player_name', 'end season', 'player_id', 'mp'], axis=1)
```

```python
all_impact_metric.to_excel('all_impact_metrics.xlsx')
```
